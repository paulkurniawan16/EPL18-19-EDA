```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('EPL_players.csv')
data.head()
```

```python
data.sample(20)
```

```python
data.info()
```

```python
del data['birthday']
pd.to_datetime(data['birthday_GMT'])
data[['rank_in_league_top_attackers',
      'rank_in_league_top_midfielders',
      'rank_in_league_top_defenders',
      'rank_in_club_top_scorer']] = data[['rank_in_league_top_attackers',
      'rank_in_league_top_midfielders',
      'rank_in_league_top_defenders',
      'rank_in_club_top_scorer']].astype(str)
```

```python
data[['rank_in_league_top_attackers',
     'rank_in_league_top_midfielders',
     'rank_in_league_top_defenders',
     'rank_in_club_top_scorer']] = data[['rank_in_league_top_attackers',
     'rank_in_league_top_midfielders',
     'rank_in_league_top_defenders',
     'rank_in_club_top_scorer']].replace('-1', np.nan)
```

```python
data.isnull().sum()
```

```python
data = data[data['minutes_played_overall'] != 0]
data = data.dropna()
data
```

```python
numericals = ['age',
              'minutes_played_overall',
              'minutes_played_home',
              'minutes_played_away',
              'appearances_overall',
              'appearances_home',
              'appearances_away',
              'goals_overall',
              'goals_home',
              'goals_away',
              'assists_overall',
              'assists_home',
              'assists_away',
              'penalty_goals',
              'penalty_misses',
              'clean_sheets_overall',
              'clean_sheets_home',
              'clean_sheets_away',
              'conceded_overall',
              'conceded_home',
              'conceded_away',
              'yellow_cards_overall',
              'red_cards_overall',
              'goals_involved_per_90_overall',
              'assists_per_90_overall',
              'goals_per_90_overall',
              'goals_per_90_home',
              'goals_per_90_away',
              'min_per_goal_overall',
              'conceded_per_90_overall',
              'min_per_conceded_overall']
data[numericals].describe()
```

```python
categoricals = ['full_name',
                'birthday_GMT',
                'league',
                'season',
                'position',
                'Current Club',
                'nationality'
                ]
data[categoricals].describe()
```

```python
for col in categoricals:
  print(f"VALUE COUNTS OF {col} COLUMN", '\n')
  print(data[col].value_counts(), '\n')
```

```python
foward = data[data['position'] == 'Forward']
foward.head()
```

```python
plt.figure(figsize=(50,24))

features = numericals
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=foward[features[i]], color='cyan')
    plt.tight_layout()
```

```python
plt.figure(figsize=(50,24))

features = numericals
for i in range(0, len(features)):
    plt.subplot(2, len(features)//2 + 1, i+1)
    #plt.subplot(1, len(features), i+1)
    sns.distplot(x=foward[features[i]], color='skyblue')
    plt.xlabel(features[i])
    plt.tight_layout()
```
